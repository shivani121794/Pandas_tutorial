```python
# Combining dataframe

import pandas as pd
```

```python
# Concate

df1 = pd.DataFrame({'Name': ['Vicky', 'Bill'], 'Age': [12,46]})
df2 = pd.DataFrame({'Name': ['John', 'Sabrina'], 'Age': [37,25]})
df1
```

```python
df2
```

```python
# Concat Horizontically

cars = pd.concat([df1, df2], axis=1)
cars
```

```python
df3 = pd.concat([df1, df2], axis=0)
df3
```

```python
# Concat Vertically and ignore index

df3 = pd.concat([df1, df2], axis=0, ignore_index=True)
df3
```

```python
df4 = pd.DataFrame({'Name':['Tyler', 'Natalie'], 'Age': [28,39], 'Hobby': ['Swimming', 'Reading']})
df4
```

```python
# Outer the default for the join argument

pd.concat([df3, df4], join = 'outer', ignore_index=True)     #shift+tab 
```

```python
pd.concat([df3, df4], ignore_index=True)
```

```python
# Inner stacks only the columns that are shared between the two dataframe

pd.concat([df3, df4], join = 'inner', ignore_index=True)
```

```python
# Merge

size = pd.DataFrame({'Color': ['Red', 'Blue', 'Green'], 'Size': ['M', 'S', 'M']})
gender = pd.DataFrame({'Color': ['Red', 'Blue', 'Yellow'], 'Sex': ['Female', 'Female', 'Male']})
size
```

```python
gender
```

```python
# Only show rows where there is a match in the column
# Red and blue are present in both dataframes
pd.merge(size,gender, how='inner')
```

```python
# Show all rows and fill rows with null values when there is no match
# There is no sex info on colour green and there is no size info on yellow colour
pd.merge(size,gender, how='outer')
```

```python
# Keep all left dataframe

pd.merge(size,gender, how='left')
```

```python
pd.merge(size,gender, how='right')
```
